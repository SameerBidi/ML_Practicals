### Non Linear Model Polynomial Regression

```python
import pandas as pd

df = pd.read_csv("datasets/PolyData.csv", index_col=0)
df.head()
```

```python
x = df.iloc[:,0:1].values
y = df.iloc[:, 1].values

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

poly.fit(x_train)

x_train_poly = poly.transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

poly_reg = LinearRegression()

poly_reg.fit(x_train_poly, y_train)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, lr.predict(x_test), color='r')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame({"x_test": x_test.flatten(), "y_pred": poly_reg.predict(x_test_poly).flatten()})
df = df.sort_values(by="x_test")

plt.scatter(x_test, y_test, color='blue')
plt.plot(df["x_test"], df["y_pred"], color='r')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
