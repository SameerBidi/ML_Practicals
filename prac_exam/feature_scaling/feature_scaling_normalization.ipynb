### Feature Scaling using Normalization

```python
import pandas as pd

df=pd.read_csv("datasets/wine_data.csv", header=None, usecols=[0, 1, 2])
df.columns = ["Class Label", "Alcohol", "Malic Acid"]
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(scaled, columns=df.columns)
df_scaled.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.scatter(df["Alcohol"], df["Malic Acid"])
ax1.set_title("Before Scaling")

ax2.scatter(df_scaled["Alcohol"], df_scaled["Malic Acid"], color="green")
ax2.set_title("After Scaling")
```

```python
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.set_title("Before Scaling")

sns.kdeplot(df["Alcohol"], ax=ax1)
sns.kdeplot(df["Malic Acid"], ax=ax1)

ax2.set_title("After Scaling")

sns.kdeplot(df_scaled["Alcohol"], ax=ax2)
sns.kdeplot(df_scaled["Malic Acid"], ax=ax2)
```
