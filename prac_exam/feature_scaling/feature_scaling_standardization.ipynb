```python
import pandas as pd

df=pd.read_csv("datasets/Social_Network_Ads.csv", usecols=["Age", "EstimatedSalary"])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(scaled, columns=df.columns)
df_scaled.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.scatter(df["Age"], df["EstimatedSalary"])
ax1.set_title("Before Scaling")

ax2.scatter(df_scaled["Age"], df_scaled["EstimatedSalary"], color="green")
ax2.set_title("After Scaling")
```

```python
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.set_title("Before Scaling")

sns.kdeplot(df["Age"], ax=ax1)
sns.kdeplot(df["EstimatedSalary"], ax=ax1)

ax2.set_title("After Scaling")

sns.kdeplot(df_scaled["Age"], ax=ax2)
sns.kdeplot(df_scaled["EstimatedSalary"], ax=ax2)
```
