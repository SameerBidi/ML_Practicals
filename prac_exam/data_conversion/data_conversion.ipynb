### TSA Data Conversion

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datasets/AirPassengers.csv")
df = df.rename(columns={"#Passengers": "Passengers"})
df
```

```python
df["Passengers"].plot()
```

#### By Differencing

```python
df['Passengers_diff1'] = df['Passengers'] - df['Passengers'].shift(1) 

df['Passengers_diff1'].plot()
```

```python
df['Passengers_diff7'] = df['Passengers'] - df['Passengers'].shift(7) 

df['Passengers_diff7'].plot()
```

#### By Transformation

```python
df["Passengers_log"] = np.log(df["Passengers"])
df["Passengers_log"].plot()
```

```python
df["Passengers_sqrt"] = np.sqrt(df["Passengers"])
df["Passengers_sqrt"].plot()
```

```python
df["Passengers_cbrt"] = np.cbrt(df["Passengers"])
df["Passengers_cbrt"].plot()
```

```python
df["Passengers_log_diff"] = df["Passengers_log"] - df["Passengers_log"].shift(1)
df["Passengers_log_diff"].plot()
```
